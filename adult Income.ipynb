```python
import pandas as pd
import numpy as np
```

```python
x=pd.read_csv("adult.csv")
```

```python
x.head()
```

```python
income_dummies=pd.get_dummies(x.income)
```

```python
income_dummies.head()
```

```python
x_dum=pd.concat([x,income_dummies],axis='columns')
```

```python
x_dum.head()
```

```python
 x_dum.rename(columns = {'<=50K':'Lessthan50K','>50K':'Morethan50K'}, inplace = True)
    
```

```python
x_dum.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_dum[['age']],x_dum[['income']],train_size=0.8,random_state=2)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_predicted = model.predict(x_test)
```

```python
model.predict_proba(x_test)
```

```python
model.score(x_test,y_test)
```
